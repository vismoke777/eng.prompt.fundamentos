```python
#ehprimo
def eh_primo(n):
  if n < 2:
    return False
  if n == 2:
    return True
  for i in range (2,n):
    if n % i == 0:
      return False
  return True
```

### Testes para a função `eh_primo`

```python
print(f"Is 2 prime? {eh_primo(2)}")
print(f"Is 7 prime? {eh_primo(7)}")
print(f"Is 10 prime? {eh_primo(10)}")
print(f"Is 17 prime? {eh_primo(17)}")
print(f"Is 25 prime? {eh_primo(25)}")
```

```python
#calculadora
import math

def calculadora(operacao, num1=None, num2=None, sugestao_funcoes=False):
    if sugestao_funcoes:
        return "Funções matemáticas sugeridas: sqrt(x) - raiz quadrada, pow(x, y) - potência, log(x, base) - logaritmo, sin(x), cos(x), tan(x)."

    if operacao == 'soma':
        if num1 is None or num2 is None: return "Erro: Soma requer dois números."
        return num1 + num2
    elif operacao == 'subtracao':
        if num1 is None or num2 is None: return "Erro: Subtração requer dois números."
        return num1 - num2
    elif operacao == 'multiplicacao':
        if num1 is None or num2 is None: return "Erro: Multiplicação requer dois números."
        return num1 * num2
    elif operacao == 'divisao':
        if num1 is None or num2 is None: return "Erro: Divisão requer dois números."
        try:
            return num1 / num2
        except ZeroDivisionError:
            return "Erro: Divisão por zero!"
    elif operacao == 'raiz_quadrada':
        if num1 is None: return "Erro: Raiz quadrada requer um número."
        if num1 < 0: return "Erro: Raiz quadrada de número negativo não é real."
        return math.sqrt(num1)
    elif operacao == 'potencia':
        if num1 is None or num2 is None: return "Erro: Potência requer dois números (base e expoente)."
        return math.pow(num1, num2)
    elif operacao == 'logaritmo':
        if num1 is None or num2 is None: return "Erro: Logaritmo requer dois números (número e base)."
        if num1 <= 0 or num2 <= 0: return "Erro: Logaritmo e base devem ser positivos."
        return math.log(num1, num2)
    else:
        return "Operação inválida. Tente 'soma', 'subtracao', 'multiplicacao', 'divisao', 'raiz_quadrada', 'potencia', 'logaritmo'. Para sugestões de funções, use sugestao_funcoes=True."
```

```python
import random

def jogar_forca():
    palavras = ['python', 'programacao', 'desenvolvimento', 'colaboratorio', 'inteligencia']
    palavra_secreta = random.choice(palavras)
    letras_certas = []
    letras_erradas = []
    tentativas_restantes = 6

    print('Bem-vindo ao Jogo da Forca!')

    while True:
        # Exibir estado atual da palavra
        palavra_oculta = ''
        for letra in palavra_secreta:
            if letra in letras_certas:
                palavra_oculta += letra
            else:
                palavra_oculta += '_'
        print(f'Palavra: {palavra_oculta}')

        # Exibir tentativas restantes e letras erradas
        print(f'Tentativas restantes: {tentativas_restantes}')
        print(f'Letras erradas: {', '.join(letras_erradas)}')

        # Verificar vitória ou derrota
        if palavra_oculta == palavra_secreta:
            print(f'Parabéns! Você adivinhou a palavra: {palavra_secreta}')
            break
        if tentativas_restantes == 0:
            print(f'Fim de jogo! A palavra era: {palavra_secreta}')
            break

        # Pedir palpite ao jogador
        palpite = input('Adivinhe uma letra: ').lower()

        if len(palpite) != 1 or not palpite.isalpha():
            print('Por favor, digite apenas uma letra.')
            continue

        if palpite in letras_certas or palpite in letras_erradas:
            print('Você já tentou esta letra.')
            continue

        if palpite in palavra_secreta:
            letras_certas.append(palpite)
            print(f'Boa! A letra {palpite} está na palavra.')
        else:
            letras_erradas.append(palpite)
            tentativas_restantes -= 1
            print(f'Ops! A letra {palpite} não está na palavra.')

# Para iniciar o jogo, chame a função:
jogar_forca()
```
