Crie um scrip python simples para verificar a temperatura de um sistema. Se for maior que 30, exiba um alerta de superaquecimento.

```python
import random

# Solicita a temperatura ao usuário
temperature_str = input("Digite a temperatura atual do sistema em °C: ")

try:
    temperature = float(temperature_str)
    print(f"Temperatura atual do sistema: {temperature:.2f}°C")

    if temperature > 30:
        print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C.")
    else:
        print("Temperatura do sistema está normal.")
except ValueError:
    print("Entrada inválida. Por favor, digite um número para a temperatura.")
```

```python
import os
import shutil

def organize_files_by_extension(source_directory):
    """
    Organiza arquivos em um diretório em subpastas baseadas em suas extensões.

    Args:
        source_directory (str): O caminho do diretório a ser organizado.
    """
    if not os.path.isdir(source_directory):
        print("Erro: O diretório '" + source_directory + "' não existe.")
        return

    print("Organizando arquivos em: " + source_directory)

    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename)

        # Ignora diretórios
        # A parte `and filename != os.path.basename(__file__)` foi removida
        # pois __file__ não é definido em ambientes interativos como Colab.
        if os.path.isfile(file_path):
            # Obtém a extensão do arquivo (ex: '.pdf', '.jpg')
            _, file_extension = os.path.splitext(filename)

            # Remove o ponto inicial da extensão para usar como nome da pasta
            # Se não houver extensão (ex: um arquivo sem ponto), usa 'no_extension'
            if file_extension:
                folder_name = file_extension[1:].lower()
            else:
                folder_name = "no_extension"

            destination_folder = os.path.join(source_directory, folder_name)

            # Cria a pasta de destino se ela não existir
            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)
                print("Criada pasta: " + destination_folder)

            # Move o arquivo para a pasta de destino
            try:
                shutil.move(file_path, destination_folder)
                print("Movido: " + filename + " -> " + destination_folder)
            except Exception as e:
                print("Erro ao mover o arquivo " + filename + ": " + str(e))
    print("Organização de arquivos concluída.")

# --- Como usar ---
# Defina o diretório que você quer organizar.
# Por padrão, ele usará o diretório atual onde o script está sendo executado.
# Se quiser organizar outro diretório, substitua '.' pelo caminho completo, ex: '/caminho/para/seus/arquivos'
source_directory_to_organize = '.' # Use '.' para o diretório atual

# Chame a função para iniciar a organização
organize_files_by_extension(source_directory_to_organize)
```

#Consulta de APIS

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço.

    Args:
        cep (str): O número do CEP a ser consultado.

    Returns:
        dict: Um dicionário com os dados do endereço, ou None se h houver erro.
    """
    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para status HTTP ruins (4xx ou 5xx)
        data = response.json()

        if 'erro' in data and data['erro']:
            print(f"Erro: CEP {cep} não encontrado.")
            return None
        else:
            return data
    except requests.exceptions.RequestException as e:
        print(f"Erro ao conectar à API ViaCEP: {e}")
        return None

def exibir_endereco_formatado(endereco):
    """
    Exibe as informações do endereço de forma legível.

    Args:
        endereco (dict): Dicionário contendo os dados do endereço.
    """
    if endereco:
        print("\n--- Detalhes do Endereço ---")
        print(f"CEP: {endereco.get('cep', 'N/A')}")
        print(f"Logradouro: {endereco.get('logradouro', 'N/A')}")
        print(f"Complemento: {endereco.get('complemento', 'N/A')}")
        print(f"Bairro: {endereco.get('bairro', 'N/A')}")
        print(f"Localidade: {endereco.get('localidade', 'N/A')}")
        print(f"UF: {endereco.get('uf', 'N/A')}")
        print(f"DDD: {endereco.get('ddd', 'N/A')}")
        print("---------------------------")
    else:
        print("Não foi possível obter os dados do endereço.")

# --- Uso do Script ---
if __name__ == "__main__":
    while True:
        cep_input = input("Digite o CEP (apenas números, ou 'sair' para encerrar): ").strip()
        if cep_input.lower() == 'sair':
            break

        if not cep_input.isdigit() or len(cep_input) != 8:
            print("CEP inválido. Por favor, digite 8 dígitos numéricos.")
            continue

        endereco_info = consultar_cep(cep_input)
        exibir_endereco_formatado(endereco_info)
```

#Sistema de notificação

```python
import datetime

def enviar_alerta_de_email_simulado(assunto, mensagem, destinatario="admin@exemplo.com"):
    """
    Simula o envio de um e-mail de alerta.
    Em um sistema real, aqui haveria a integração com um serviço de e-mail (SMTP).
    """
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"\n--- ALERTA DE E-MAIL SIMULADO ({timestamp}) ---")
    print(f"De: sistema@exemplo.com")
    print(f"Para: {destinatario}")
    print(f"Assunto: {assunto}")
    print(f"\n{mensagem}")
    print(f"------------------------------------------------\n")

def verificar_condicao_critica(valor_monitorado, limite_critico, nome_do_monitoramento):
    """
    Verifica se um valor monitorado atingiu uma condição crítica.
    Se sim, dispara alertas.
    """
    if valor_monitorado > limite_critico:
        # Alerta no console
        print(f"ALERTA CRÍTICO NO CONSOLE: {nome_do_monitoramento} ({valor_monitorado}) excedeu o limite de {limite_critico}!")

        # Alerta via e-mail simulado
        assunto_email = f"ALERTA CRÍTICO: {nome_do_monitoramento} Elevado"
        mensagem_email = (
            f"Prezado(a) administrador(a),\n\n"
            f"O sistema detectou uma condição crítica para '{nome_do_monitoramento}'.\n"
            f"Valor atual: {valor_monitorado}\n"
            f"Limite crítico: {limite_critico}\n\n"
            f"Por favor, investigue imediatamente.\n\n"
            f"Atenciosamente,\n"
            f"Sistema de Monitoramento"
        )
        enviar_alerta_de_email_simulado(assunto_email, mensagem_email)
        return True
    else:
        print(f"Status normal para {nome_do_monitoramento}: {valor_monitorado}°C (limite: {limite_critico}°C).")
        return False

# --- Exemplos de Uso ---
if __name__ == "__main__":
    print("\n--- Iniciando monitoramento de exemplo ---")

    # Exemplo 1: Temperatura normal
    print("\n*** Cenário: Temperatura Normal ***")
    temp_atual_1 = 28.5
    limite_temp = 30.0
    verificar_condicao_critica(temp_atual_1, limite_temp, "Temperatura do Sistema")

    # Exemplo 2: Temperatura crítica
    print("\n*** Cenário: Temperatura Crítica ***")
    temp_atual_2 = 35.2
    verificar_condicao_critica(temp_atual_2, limite_temp, "Temperatura do Sistema")

    # Exemplo 3: Outra condição crítica (e.g., uso de CPU)
    print("\n*** Cenário: Uso de CPU Crítico ***")
    uso_cpu_atual = 92.0
    limite_cpu = 90.0
    verificar_condicao_critica(uso_cpu_atual, limite_cpu, "Uso da CPU")

    print("\n--- Monitoramento de exemplo concluído ---")
```
